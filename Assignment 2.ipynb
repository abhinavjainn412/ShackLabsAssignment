```python
import pandas as pd
```

```python
df1=pd.read_csv('amz_com-ecommerce_sample.csv',encoding='latin-1')
```

```python
df1.head()
```

```python
df2=pd.read_csv('flipkart_com-ecommerce_sample.csv',encoding='latin-1')
```

```python
df2.head()
```

```python
# taking the required columns from the dataset
df11=df1[['uniq_id','product_name','retail_price','discounted_price']]
```

```python
df22=df2[['uniq_id','product_name','retail_price','discounted_price']]
```

```python
df11.head()
```

```python
df22.head()
```

```python
# merging amazon and flipkart dataset
df_final=pd.merge(left = df11, right = df22, how = 'inner', on = 'uniq_id' )
```

```python
df_final.head()
```

```python
df_final=df_final.drop(['uniq_id'],axis=1)
```

```python
df_final=df_final.rename(columns={'product_name_x':'Product name in Amazon','retail_price_x':'Retail Price in Amazon','discounted_price_x':'Discounted Price in Amazon','product_name_y':'Product name in Flipkart','retail_price_y':'Retail Price in Flipkart','discounted_price_y':'Discounted Price in Flipkart'})
```

```python
df_final.head(n=10)
```

```python
### The final dataset is converted to csv file. 
df_final.to_csv('df_final.csv')
```
